# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from collections import Counter
import string
from datetime import datetime


from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, silhouette_score
from scipy.stats import ttest_ind
from gensim.models import Word2Vec
from sklearn.manifold import TSNE


%matplotlib inline

# Options for pandas
pd.options.display.max_columns = 150
pd.options.display.max_rows = 150
```

[kaggle dataset](https://www.kaggle.com/carrie1/ecommerce-data)

```python
df = pd.read_csv('data/data.csv')
```

```python
df.head()
```

# Data Cleaning

```python
df = df[df['UnitPrice'] != 0]
df['cancel'] = df['InvoiceNo'].apply(lambda x: 1 if re.match('^[a-zA-Z]+', x) else 0)
df = df[df['cancel'] == 0]

product_df = df[['StockCode', 'Description', 'UnitPrice']]


remove_stockcode_list = [ 'D', 'C2', 'DOT', 'M', 'BANK CHARGES','S','AMAZONFEE','m','gift_0001_50','gift_0001_30',
                         'gift_0001_20','gift_0001_10','PADS','B','CRUK']
product_df = product_df[~product_df['StockCode'].isin(remove_stockcode_list)]

product_df = product_df.groupby('StockCode').first()
```

# Feature Engineering

```python
product_df['word_list'] = product_df['Description'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)).strip().lower().split())
```

```python
word_embeddings = Word2Vec(product_df['word_list'].tolist(), min_count=1)
```

```python
product_df['vector'] = product_df['word_list'].apply(lambda word_list: sum([word_embeddings.wv[word] for word in word_list]) / len(word_list))
```

```python
df['datetime'] = pd.to_datetime(df['InvoiceDate'])
df['month'] = df['datetime'].apply(lambda x: x.month)
df['day'] = df['datetime'].apply(lambda x: x.day)
```

# Clustering

```python
product_df.head()
```

```python
vectors = np.vstack((product_df['vector'])).reshape(100,-1)

unit_price = np.array(product_df['UnitPrice']).reshape(1,-1)

data = np.vstack((vectors,unit_price)).T
data = normalize(data)
```

```python
n_clusters = 60
inertia = {}
silhouette = {}
for i in range(2, n_clusters):
    km = KMeans(n_clusters=i, random_state=42)
    y_pred = km.fit_predict(data)
    inertia[i] = km.inertia_
    silhouette[i] = silhouette_score(data, y_pred, metric='euclidean')
    
plt.plot(inertia.keys(), inertia.values())
plt.xticks(np.arange(0,max(list(inertia.keys())),2))
plt.xlabel('N Clusters')
plt.ylabel('Sum-Squared Difference')
plt.title('Cluster Inertia')
plt.show()

plt.plot(silhouette.keys(), silhouette.values())
plt.xticks(np.arange(0,max(list(inertia.keys())),2))
plt.xlabel('N Clusters')
plt.ylabel('Sum-Squared Difference')
plt.title('Cluster Silhouette')
plt.show()
```

```python
km = KMeans(n_clusters=8, random_state=42)
km_pred = km.fit_predict(data)
clusters = km.cluster_centers_
```

```python
bandwidth = estimate_bandwidth(data)/2
ms = MeanShift(bandwidth=bandwidth, cluster_all=True).fit(data)
ms_clusters = len(ms.cluster_centers_)
ms_pred = ms.labels_

print('Bandwidth: ', bandwidth)
print('N Clusters: ', ms_clusters)
```

```python
tsne = TSNE(n_components=2, n_iter=500)
tsne_results = tsne.fit_transform(data)
tsne_results = tsne_results.T
```

```python
plt.scatter(tsne_results[0], tsne_results[1], c=ms_pred, alpha=.1)
plt.show()
```

```python
plt.hist(ms_pred, bins=50)
plt.xlabel('Cluster Label')
plt.ylabel('Count')
plt.title('Cluster Distribution')
plt.show()
```

```python
plt.hist(km_pred, bins=50)
plt.xlabel('Cluster Label')
plt.ylabel('Count')
plt.title('Cluster Distribution')
plt.show()
```
