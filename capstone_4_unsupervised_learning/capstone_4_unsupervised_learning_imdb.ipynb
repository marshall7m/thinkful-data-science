# Feature Engineering

## Word Embedding

Convert movie descriptions into word embeddings:
   - Lowercase
   - Split at spaces 
   - Remove punctuations and stops words
   - Filter out words that have a total count of less than five within the entire dataset description word count
   - Transform into word vectors (vector length = 100)
   - Final vector is the average of all word vectors for each respective movie
   

```python
df['word_list'] = df['Description'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)).strip().lower().split())

stop_words = list(stopwords.words())

def remove_stopwords(word_list):
    filtered_word_list = [word for word in word_list if word not in stop_words]
    return filtered_word_list

df['clean_word_list'] = df['word_list'].apply(remove_stopwords)

word_embeddings = Word2Vec(df['clean_word_list'].tolist(), min_count=5)
```

```python
word_embeddings.wv.similar_by_word('fight')
```

```python
def vector_funct(word_list):
    filter_word_list = [word for word in word_list if word in word_embeddings]
    if len(filter_word_list) != 0:
        return sum([word_embeddings.wv[word] for word in filter_word_list ]) / len(filter_word_list)
    else:
        return np.zeros(100)


df['vector'] = df['clean_word_list'].apply(vector_funct)

```

## Target Variable: Key Genre 

```python
def get_key_genre(genre_list):
    genre_list = genre_list.split(',')
    key_genre_list = [genre for genre in genre_list if genre in key_genre]
    if len(key_genre_list) == 1:
        return key_genre_list[0]
    
    elif len(key_genre_list) > 1:
        return 'multiple'
    
    else:
        return 'NA'
            

df['Genre'].value_counts()
key_genre = ['Action', 'Drama', 'Comedy']

df['key_genre'] = df['Genre'].apply(get_key_genre)
df['key_genre'].value_counts()
```

## One-Hot Encoded Words

- Collect top five most frequent description words within each key genre
- If that genre's frequent word count is twice the count of another key genre then that words will be an one-hot encoded feature


```python
#get all clean words from description column
all_words = [word for word_list in df['clean_word_list'] for word in word_list]

#get frequency of words
all_words_dict = dict(Counter(all_words))

genre_list = list(df['key_genre'].unique())

top_genre_word_dict = {}
genre_word_dict = {}
dummy_words = []

for genre in genre_list:
    
    #get all words within genre
    genre_words = [word for word_list in df['clean_word_list'][df['key_genre'] == genre] for word in word_list]
    genre_word_dict[genre] = dict(Counter(genre_words))
    
    #get top 5 most frequent words within respective genre
    top_genre_words = dict(Counter(genre_words).most_common()[:20])
    
    #add top words to genre_word_dict
    top_genre_word_dict[genre] = top_genre_words
    
    
    x_genre = genre
    #get list of other genre for ratios
    other_genres = [genre for genre in genre_list if genre != x_genre]
    for word in top_genre_word_dict[x_genre].keys():
        for genre in other_genres:
            try:
                other_genre_freq = genre_word_dict[genre][word]
            except KeyError:
                other_genre_freq = 0
            
            x_genre_freq = top_genre_word_dict[x_genre][word]
            
            ratio = other_genre_freq / x_genre_freq
            if ratio < .9:
                dummy_words.append(word)
                
dummy_words = list(np.unique(dummy_words))
```

# Clustering

Separate supervised data from unsupervised data

Supervised Data:
    - Explanatory include continous variables, dummy variables and word embeddings
    - Movies with multiple key genres are filtered out
        - Models will have a hard time distinguishing between two movies such as a action and drama movie vs. action movie
        - A movie with multiple key genre will have a large similarity with a movie that has a single related key genre
    
Unsupervised data:
    - Explanatory include continous variables, dummy variables and word embeddings
    - Includes movies that fall under multiple key genres since clustering based on genres isn't explicit in unsupervised models and more data the better

```python
standard_s = StandardScaler()

continous_features = ['winsorized_Revenue (Millions)', 'winsorized_Votes', 'Rating', 'Metascore', 'Runtime (Minutes)']

cluster_vectors = np.vstack((df['vector']))
cluster_feature_df = df[dummy_cols + continous_features]
cluster_data = np.hstack((cluster_vectors, cluster_feature_df))

cluster_normalize_data = standard_s.fit_transform(cluster_data)

classification_df = df[df['key_genre']!= 'multiple'].copy()
classification_feature_df = classification_df[dummy_cols + continous_features]
classification_vectors = np.vstack((classification_df['vector']))
classification_data = np.hstack((classification_vectors, classification_feature_df))

classification_normalize_data = standard_s.fit_transform(classification_data)
```

```python
def plot_tsne_clusters(ap_pred, sc_pred, ms_pred, km_pred, tsne_results, skip_ap=False):
    plt.figure(figsize=(12,8))

    if skip_ap == True:
        plt.subplot(2,2,1)
        plt.title('Spectral')
        plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=sc_pred, alpha=.5)

        plt.subplot(2,2,2)
        plt.title('Mean Shift')
        plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=ms_pred, alpha=.5)

        plt.subplot(2,2,3)
        plt.title('K-Means')
        plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=km_pred, alpha=.5)
    
    else:
        plt.subplot(2,2,1)
        plt.title('Affinity Propagation')
        plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=ap_pred, alpha=.5)

        plt.subplot(2,2,2)
        plt.title('Spectral')
        plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=sc_pred, alpha=.5)

        plt.subplot(2,2,3)
        plt.title('Mean Shift')
        plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=ms_pred, alpha=.5)

        plt.subplot(2,2,4)
        plt.title('K-Means')
        plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=km_pred, alpha=.5)

    plt.tight_layout()
    plt.show()
```

# Supervised Modeling

Grid Search CV uses the dictionary of hyperparameters to return the best combination of hyperparameters for each model.

```python
rfc = RandomForestClassifier()
rfc.parameters = {'criterion':('gini', 'entropy'), 'n_estimators':[5,10,20,40], 
                  'max_depth': [1,2,4,8,16,32], 'min_samples_leaf': [1,2,4,8]}

dt = DecisionTreeClassifier()
dt.parameters = {'criterion':('gini', 'entropy'), 'max_depth': [1,2,4,8,16,32], 'min_samples_leaf': [1,2,4,8]}

knn = KNeighborsClassifier()
knn.parameters = {'weights':('uniform', 'distance'), 'n_neighbors':[1,2,4,8]}

mlp = MLPClassifier()
mlp.parameters = {'hidden_layer_sizes':[10,15,20], 'activation':('relu', 'logistic'), 
              'learning_rate': ('constant', 'adaptive')}

gb = GradientBoostingClassifier()
gb.parameters = {'learning_rate': [.001,.01, .1], 'criterion': ('friedman_mse', 'mae')}


model_dict = {'rfc': rfc, 'mlp':mlp, 'knn':knn, 'dt': dt}
```

```python
for name, model in model_dict.items():
    clf = GridSearchCV(model, model.parameters, cv=5, scoring='accuracy')
    if name in ['mlp', 'knn']:
        clf.fit(classification_normalize_data, classification_df['key_genre'])
    else:
        clf.fit(classification_data, classification_df['key_genre'])

    model_dict[name] = clf.best_estimator_
model_dict
```

```python
for name,model in model_dict.items():
    if name in ['knn', 'mlp']:
        train_x, test_x, train_y, test_y = train_test_split(classification_normalize_data, classification_df['key_genre'], test_size=.3)
    else:
        train_x, test_x, train_y, test_y = train_test_split(classification_data, classification_df['key_genre'], test_size=.3)
    
    start = time.time()
    model.fit(train_x, train_y)
    end = time.time()
    train_score = model.score(train_x, train_y) * 100
    score = model.score(test_x, test_y) * 100
    if  name == 'rfc':
        importances = model.feature_importances_
    
    print()
    print(name)
    print('Runtime: ', end - start)
    print('Train Accuracy: ', train_score)
    print('Test Accuracy: ', score)

```
