### Imports

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline

# Options for pandas
pd.options.display.max_columns = 150
pd.options.display.max_rows = 150
```

```python
df = pd.read_csv('data/WELLCOME_APCspend2013_forThinkful.csv', encoding='iso-8859-1')
```

```python
df.dropna()
```

```python
df['Journal title'] = df['Journal title'].apply(lambda x: str(x).lower().strip())
```

```python
top_journals = df['Journal title'].value_counts()[:5]
print('Top 5 Journals and their total article count:\n', top_journals)
```

```python
pattern = re.compile('\$|\£')

df['cost'] = df['COST (£) charged to Wellcome (inc VAT when charged)'].apply(lambda x: float(pattern.sub('', x).strip()))
```

```python
df.drop(df.loc[df['cost'] == 999999.00].index, inplace=True)
```

```python
top_df = df[df['Journal title'].isin(top_journals.keys())]

top_df.groupby('Journal title').agg([np.mean, np.median, np.std])
```
