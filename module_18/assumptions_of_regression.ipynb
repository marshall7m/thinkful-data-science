````python
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))

weather_df = pd.read_sql_query('select * from weatherinszeged',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()
````

````python
#pct of nulls in columns
weather_df.isnull().sum()/weather_df.isnull().count()*100
````

````python
X = weather_df[['humidity', 'windspeed', 'windbearing', 'pressure']]
Y = weather_df['temperature']

model = LinearRegression()

model.fit(X,Y)

print("Model's Coefficients: ", model.coef_)
print("Model's Intercept: ", model.intercept_)
````

````python
pred_y = model.predict(X)

plt.scatter(Y, pred_y, alpha=.3)
plt.xlabel('Target')
plt.ylabel('Predicted Target')
````

````python
import statsmodels.api as sm

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
````

### Assumption two: the error term should be zero on average

````python
error_term = Y - pred_y
error_term.mean()
````

The average error term is below zero. This would mean that on average the model is predicting above the target temperature IF there was no constant term. Since there is a constant term the error term should be close to zero on average.

### Assumption three: homoscedasticity

````python
plt.scatter(pred_y, error_term)
plt.xlabel('Predicted')
plt.ylabel('Residual')
plt.axhline(y=0)
plt.title('Residual vs. Predicted')
plt.show()
````

Visually speaking, it seems like the higher prediction have more error variance compared to lower predictions. 

````python
#testing for heteroscedasticity in error term

from scipy.stats import bartlett
from scipy.stats import levene

bartlett_test = bartlett(pred_y, error_term)
levene_test = levene(pred_y, error_term)

print(bartlett_test)
print(levene_test)
````

Using both the bartlett and levene test we can determine if the error term distribution is homoscedasticity or heteroscedasticity. Since pval > .05 for both test we can not reject the null which means the the distribution is homoscedasticity and the error term variance is consistent throughout the range of observations. 

### Assumption four: low multicollinearity

The coefficient correlation between the independent variables are weak. This means that there is a low multicollinearity.

````python
plt.plot(error_term)
````

````python
acf_data = acf(error_term)

plt.plot(acf_data[1:])
plt.show()
````

````python
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'houseprices'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
housing_df = pd.read_sql_query('select * from houseprices',con=engine)

engine.dispose()
````

````python
#create dummy variables for categorical features

for col in housing_df.columns:
    #create dummy variables for binary features
    if (housing_df[col].dtype == 'O') and (len(housing_df[col].value_counts()) == 2):
        binary_col = housing_df[col].value_counts().keys()[0]
        housing_df["is_" + binary_col] = pd.get_dummies(housing_df[col], drop_first=True)
    #create dummy variables for other multi-category features
    elif (housing_df[col].dtype == 'O') and (len(housing_df[col].value_counts()) != 2):
        housing_df = pd.concat([housing_df, pd.get_dummies(housing_df[col], prefix=col)], axis=1)
````

````python
corr = housing_df.corr()
plot_corr = corr[(corr.saleprice > .5) | (corr.saleprice < -.5)]
plot_corr = plot_corr.saleprice
plot_corr.drop('saleprice', inplace = True)

feature_list = list(plot_corr.keys())

# remove one feature from pairs of multicollinearity features (see linear_regression.ipynb)

remove_from_list = ['garagecars', 'totrmsabvgrd', 'firstflrsf']
for col in remove_from_list:
    feature_list.remove(col)
````

````python
feature_list
````

````python
model = LinearRegression()

X = housing_df[feature_list]
Y = housing_df['saleprice']

model.fit(X, Y)

pred_y = model.predict(X)

error_term = Y - pred_y

#testing for heteroscedasticity in error term

bartlett_test = bartlett(pred_y, error_term)
levene_test = levene(pred_y, error_term)

print('\nCoefficients: \n', model.coef_)
print('\nIntercept: \n', model.intercept_)
print('\nError Term Mean: \n', error_term.mean())
print('\n', bartlett_test)
print(levene_test)
````

````python
X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
````

````python
plt.scatter(X.overallqual, pred_y)
plt.xlabel("feature")
plt.ylabel("target")
plt.title('Linear regression can capture quadratic relationship')
plt.show()
````
