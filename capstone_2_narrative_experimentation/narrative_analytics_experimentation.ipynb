```python
df['Timestamp'] = pd.to_datetime(df.Timestamp)
df['Hour'] = df.Timestamp.dt.hour
```

```python
# separate into clicked on ad group and not clicked on ad group

clicked_group = df[df['Clicked on Ad'] == 1]
not_clicked_group = df[df['Clicked on Ad'] == 0]

print('Not Clicked Group Size: {}'.format(len(not_clicked_group)))
print('Clicked Group Size: {}'.format(len(clicked_group)))
```

```python
plt.figure(figsize=(12,20))
plt.subplot(4,2,1)
sns.boxplot(x='Male', y='Age', data=not_clicked_group)
plt.title("Didn't click on Ad")

plt.subplot(4,2,2)
sns.boxplot(x='Male', y='Age', data=clicked_group)
plt.title("Clicked on Ad")

plt.subplot(4,2,3)
sns.boxplot(x='Male', y='Daily Internet Usage', data=not_clicked_group)
plt.title("Didn't click on Ad")

plt.subplot(4,2,4)
sns.boxplot(x='Male', y='Daily Internet Usage', data=clicked_group)
plt.title("Clicked on Ad")

plt.subplot(4,2,5)
sns.boxplot(x='Male', y='Daily Time Spent on Site', data=not_clicked_group)
plt.title("Didn't click on Ad")

plt.subplot(4,2,6)
sns.boxplot(x='Male', y='Daily Time Spent on Site', data=clicked_group)
plt.title("Clicked on Ad")

plt.subplot(4,2,7)
sns.boxplot(x='Male', y='Area Income', data=not_clicked_group)
plt.title("Didn't click on Ad")

plt.subplot(4,2,8)
sns.boxplot(x='Male', y='Area Income', data=clicked_group)
plt.title("Clicked on Ad")

plt.tight_layout()
plt.show()
```

```python
sns.distplot(clicked_group.Hour, bins=50, kde=False)
plt.xticks(np.arange(0, max(clicked_group.Hour)+1))
plt.ylabel('Frequency')
plt.title('Clicked on Ad Hours')

plt.show()
```

```python
sns.scatterplot(x='Area Income', y='Daily Internet Usage', hue='Male', alpha=.7, legend=False, data=clicked_group)
plt.legend(title='Gender', labels=['Female','Male'])
```

# Proposal

According to an [interview](https://www.emarketer.com/Article/Millennials-Adept-Filtering-Ads/1012335) with Chancellor Professor of Marketing at University of Massachusetts Dartmouth, Nora Ganim Barnes has found that the Millenial generation is less proned to traditional advertising. This is prevalent within the dataset. The average age of the people who didn't click on the ad is 32 years old. The average age of the people who clicked on the ad is 40 years old. Even though Millenials are less enticed by ads, different genders are interested in different categories of ads in all age groups. Female are more interested in beauty and health advertisments and males are more interested in gadgets and technology. In the dataset, the ads were not focused on specific genders. In this ad experiment, we will spefically target different genders based upon the categories above.

## Hypothesis

The gender specific ad group will have a higher ad click count than the original random ad group.

## Rollout Plan

Ad content:

Males: Latest technology gadgets

Females: Trendy fashion and beauty products

Advertisment Content Requirements:

    - Products must be advertised at full price

    - Advertisments must be sized the same

    - Advertisments must be pictures (no animations)

    - Content of advertisments will not include celebrities

__Sampling__:

Each user who visits the site will have a 50% chance of being shown the gender specific ad or the original random ad. 

__Sample sizes__:

Group A: 500 - Group B: 500

__Sucess Metric__:

Clicked on ads count in each group

__Second Metric__: 

Average age in gender targeted group

__Controlled variables__:
User will only be shown one ad 

Content of the site will be the same in both groups

__Experiment Time Range__:

Experiment will end when 500 records are documented in each group. 

## Evaluation Plan

Calculate success metrics for both groups. Determine if there's a significant difference between both groups using a t-test. P-value threshold  = (p < .05)

Outcomes:

If the gender targeted group has a higher count of clicked ads then implement the gender specific ads into orignal website. Otherwise, keep original ads on website.
